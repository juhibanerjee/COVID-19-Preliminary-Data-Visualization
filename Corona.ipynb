```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import os
data=pd.read_csv("corona_final.csv")
data.shape
```

```python
data.head()
```

```python
data
```

```python
data.describe()
```

```python
f, ax = plt.subplots(figsize=(20,5))
sns.barplot(x="Total Cases", y="Total Deaths", data=data, ax=ax)
plt.xticks(rotation=60)
plt.xlabel('Total Cases', fontsize=15)
plt.ylabel('Total Deaths',fontsize=15)
plt.show()
```

```python
corr = data.corr()
sns.heatmap(corr)
plt.show()
```

```python
sns.set_style('whitegrid')
f, ax = plt.subplots(figsize=(20,8))
sns.boxplot(x='Total Recovered',y='Total Deaths',data=data,ax=ax)
plt.show()
```

```python
sns.lmplot(x='Total Deaths', y='Total Recovered',data=data)
plt.show()
```

```python
data.sort_values(["Total Cases", "Country"], axis=0, ascending=False, inplace=True)
data.head(20)
```

```python
f, ax = plt.subplots(figsize=(20,15))
sns.barplot(x="Country", y="Total Deaths", data=data.head(20),ax=ax)
plt.xticks(rotation=60)
plt.ylabel('Number of deaths',fontsize=20)
plt.xlabel('Countries',fontsize=20)
plt.show()
```

```python
f, ax = plt.subplots(figsize=(20,15))
sns.barplot(x="Country", y="Total Cases", data=data.head(20),ax=ax)
plt.xticks(rotation=60)
plt.ylabel('Number of deaths',fontsize=20)
plt.xlabel('Countries',fontsize=20)
plt.show()
```

```python
plt.rcParams['figure.figsize']=(20,20)
data=data.head(20)
data.groupby(['Country']).sum().plot(kind='bar')
plt.xticks(rotation=60)
plt.ylabel('Number of cases',fontsize=10)
plt.xlabel('Country',fontsize=10)
plt.show()
```
